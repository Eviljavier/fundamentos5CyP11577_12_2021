#La palabra reservada de none de python
Esta palabra reservada significa vacio,equivale a la palabra reservada void del lenguaje c , c++ y java.

Se usa para indicar ya sea un return vacio o que se recive como parametro vacio

```python
def multipicar(valor,veces):
    if valor == None:
        c=-1
    else:
            c =valor*veces
    return c
```

```python
res = multiplicar(None,5)
print(res)
```

#Argumento por defecto
Es la capacidad de las funciones de establecer un valor por defecto
Si no se envia un argumento a dicho parametro, entonces se asigna el que el programador asigno como default 

```python
def dividir(op1,op2=1):
    c = op1 / op2
    return c
print(dividir(25))
```

#argumetos enviados en un diccionario
Es el mismo concepto que el anterior pero se usa el comodin** y se mapea a 

```python
def comanda3( **opciones ):
    ops = opciones.keys()
    print(ops)
    for key in ops:
        print(f"{key} = {opciones[key]}")
    print(f"el comensal{opciones['comensal']}pidio:")

comanda3(segundo="arroz blanco",tercero="Esparragos al horno",primer="Ensalada",comensal=3)
```
